## 1 - Why Documentation Matters

### The Problem

Groundwater models often outlive their creators' memory of the details. Common scenarios:

- **"What did we assume for recharge?"** - No record of parameter choices
- **"Which version was used for the report?"** - Multiple model folders, unclear which is final
- **"Why was this boundary set here?"** - Decision rationale lost
- **"Can we update the model with new data?"** - Original workflow unclear

### The Solution

Systematic documentation serves three audiences:

| Audience | Needs | Documentation Type |
|----------|-------|-------------------|
| **Future you** | Remember decisions, reproduce results | Decision log, version control |
| **Colleagues** | Understand and modify the model | Technical documentation |
| **Stakeholders** | Trust and use results | Model report, metadata |

### Professional Standards

Major guidelines require documentation:

- **Australian Groundwater Modelling Guidelines**: Detailed reporting requirements
- **ASTM D5718**: Standard guide for documenting groundwater models
- **USGS Guidelines**: Comprehensive model archiving standards

## 2 - Version Control with Git

### Why Version Control?

Version control tracks every change to your model files, enabling:

- **History**: See what changed, when, and why
- **Rollback**: Return to any previous state
- **Collaboration**: Multiple people can work on the same model
- **Branches**: Test changes without affecting the main model

### Git Basics for Modellers

```bash
# Initialize a repository
git init

# Track changes
git add model_files/
git commit -m "Initial model setup with K=15 m/day"

# Create a branch for testing
git checkout -b test_new_recharge

# View history
git log --oneline
```

### What to Track

| Track (text files) | Don't Track (large/binary) |
|-------------------|---------------------------|
| Input files (.nam, .dis, etc.) | Output files (.hds, .cbc) |
| Scripts (.py, .ipynb) | Large datasets |
| Documentation (.md, .txt) | External data (use links) |
| Configuration files | Temporary files |

### Commit Message Best Practices

Good commit messages explain **what** and **why**:

```
# Good
"Increase K in zone 2 from 10 to 25 m/day based on pump test results"
"Add river boundary for Sihl based on gauge data 2015-2020"

# Bad
"Updated model"
"Changes"
"Fixed stuff"
```

## 3 - Model Documentation Standards

### Essential Documentation Components

#### 1. Model Metadata

Basic information that should accompany every model:

```yaml
# model_metadata.yaml
model_name: Limmat Valley Aquifer Model
version: 1.2.0
date_created: 2024-01-15
date_modified: 2024-11-20
author: Your Name
organization: ETH Zürich
purpose: Educational demonstration of groundwater flow
software: MODFLOW-NWT via FloPy
status: Calibrated, not validated (insufficient data)
```

#### 2. Decision Log

Record significant modelling decisions:

| Date | Decision | Rationale | Impact |
|------|----------|-----------|--------|
| 2024-01-15 | Single layer model | Aquifer is relatively homogeneous | Simplifies calibration |
| 2024-02-01 | K = 15 m/day initial | Literature value for gravel | Starting point for calibration |
| 2024-03-10 | No-flow at valley sides | Bedrock boundary | Consistent with geology |

#### 3. Data Sources

Document where data came from:

| Data Type | Source | Date Retrieved | Quality Notes |
|-----------|--------|----------------|---------------|
| Topography | Swisstopo DHM25 | 2024-01-10 | 25m resolution |
| Geology | Geological Atlas 1:25000 | 2024-01-10 | Some interpolation |
| GW Levels | AWEL monitoring network | 2024-02-01 | 3 wells only |

#### 4. Parameter Summary

Final calibrated parameters with justification:

| Parameter | Value | Unit | Source/Justification |
|-----------|-------|------|---------------------|
| K (gravel) | 17.7 | m/day | Calibrated (literature: 10-100) |
| Recharge | 99 | mm/year | ~10% of precipitation |
| River conductance | 100 | m²/day | Estimated from riverbed |

#### 5. Limitations Statement

Be explicit about what the model cannot do:

```markdown
## Model Limitations

1. **Validation**: Model is unvalidated due to limited observation data (n=3)
2. **Temporal**: Steady-state only - cannot predict transient responses
3. **Spatial**: Single layer - vertical flow not represented
4. **Uncertainty**: Parameter uncertainty not formally quantified

## Appropriate Uses
- Screening-level analysis
- Conceptual understanding
- Educational purposes

## Inappropriate Uses
- Regulatory decisions
- Detailed design
- Quantitative predictions without additional validation
```

## 4 - Update and Maintenance Protocols

### When to Update a Model

Triggers for model updates:

| Trigger | Action Required |
|---------|----------------|
| New observation data | Recalibrate, revalidate |
| New stress (pumping well) | Add stress, re-run scenarios |
| Boundary condition change | Update boundaries, recalibrate if needed |
| Software update | Test compatibility, document changes |
| Error discovered | Fix, document, re-run affected analyses |

### Versioning Scheme

Use semantic versioning (MAJOR.MINOR.PATCH):

- **MAJOR** (1.0 → 2.0): Fundamental model changes (new boundaries, layers)
- **MINOR** (1.0 → 1.1): Significant updates (recalibration, new stresses)
- **PATCH** (1.0.0 → 1.0.1): Bug fixes, minor corrections

### Update Checklist

When updating a model:

1. ☐ Create a new branch (Git)
2. ☐ Document the reason for update
3. ☐ Make changes
4. ☐ Re-run calibration/validation if needed
5. ☐ Update documentation
6. ☐ Review changes
7. ☐ Merge and tag new version
8. ☐ Archive previous version

### Responsibility Assignment

For team projects, clarify roles:

| Role | Responsibility |
|------|---------------|
| Model Owner | Overall quality, major decisions |
| Data Manager | Input data updates, quality control |
| Reviewer | Check changes before acceptance |
| Archivist | Long-term storage and access |

## 5 - Archiving for Reproducibility

### What to Archive

A complete model archive includes:

```
model_archive_v1.2.0/
├── README.md              # Overview and quick start
├── model_metadata.yaml    # Version, author, purpose
├── model_files/           # MODFLOW input files
│   ├── model.nam
│   ├── model.dis
│   └── ...
├── scripts/               # Processing scripts
│   ├── run_model.py
│   └── postprocess.py
├── data/                  # Input data (or links)
│   ├── observations.csv
│   └── boundary_data.csv
├── documentation/         # Reports and logs
│   ├── model_report.pdf
│   └── decision_log.md
└── environment.yml        # Software dependencies
```

### Archive Locations

| Location | Use Case | Retention |
|----------|----------|----------|
| GitHub/GitLab | Active development, collaboration | Ongoing |
| Institutional server | Official project archives | 10+ years |
| Zenodo/Figshare | Public, citable archives | Permanent |
| Local backup | Working copies | As needed |

### Reproducibility Checklist

Before archiving, verify:

1. ☐ Can someone else run the model from your archive?
2. ☐ Are all dependencies documented (environment.yml)?
3. ☐ Is the software version specified?
4. ☐ Are data sources documented or included?
5. ☐ Is there a README with instructions?

## 6 - Advanced Topics

### Automated Documentation

Tools to generate documentation automatically:

- **FloPy's `model.check()`**: Identifies common model issues
- **Jupyter notebooks**: Combine code, results, and narrative
- **Sphinx/MkDocs**: Generate documentation websites from markdown

### Continuous Integration

Automate testing when changes are made:

```yaml
# .github/workflows/model_check.yml
name: Model Check
on: [push]
jobs:
  test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v2
      - name: Run model check
        run: python scripts/check_model.py
```

### FAIR Principles for Models

Making models **F**indable, **A**ccessible, **I**nteroperable, **R**eusable:

| Principle | Implementation |
|-----------|---------------|
| **Findable** | DOI, metadata, searchable repositories |
| **Accessible** | Open access where possible, clear licensing |
| **Interoperable** | Standard formats, documented interfaces |
| **Reusable** | Clear licensing, comprehensive documentation |

### Model Registries

Emerging platforms for model sharing:

- **HydroShare**: Hydrological model and data sharing
- **CUAHSI**: Community hydrologic data services
- **ModelMy Watershed**: Web-based model access

### Long-Term Preservation

Considerations for models that must last decades:

- Format obsolescence (prefer open formats)
- Software dependencies (containerization with Docker)
- Institutional memory (comprehensive documentation)
- Legal requirements (data retention policies)

## 7 - Summary Checklist

### Minimum Documentation

Every model should have:

| Item | Purpose |
|------|--------|
| ☐ README.md | Overview, quick start, contact |
| ☐ Model metadata | Version, date, author, status |
| ☐ Parameter summary | Calibrated values with justification |
| ☐ Data sources | Where data came from |
| ☐ Limitations | What the model cannot do |
| ☐ Version control | Git or equivalent |

### Professional Projects

Add for consulting/regulatory work:

| Item | Purpose |
|------|--------|
| ☐ Decision log | Rationale for choices |
| ☐ Model report | Comprehensive technical documentation |
| ☐ Quality assurance | Review records |
| ☐ Archive plan | Long-term storage strategy |
| ☐ Update protocol | When/how to maintain |

### Key Messages

1. **Document as you go** - It's much harder to reconstruct later
2. **Use version control** - Git is the standard for a reason
3. **Think of future users** - Including future you
4. **Be explicit about limitations** - Transparency builds trust
5. **Archive complete packages** - Everything needed to reproduce results
