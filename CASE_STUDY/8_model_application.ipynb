## 1 - Introduction: From Calibration to Application

### The Purpose of Modelling

Models are built to answer questions. The entire modelling process—from conceptualization through calibration—serves this ultimate purpose. Model application is where the investment pays off.

### Types of Model Applications

| Application Type | Description | Example |
|-----------------|-------------|----------|
| **Predictive simulation** | Forecast future conditions | Head levels in 10 years |
| **Scenario testing** | Explore "what-if" situations | Impact of new pumping well |
| **Sensitivity analysis** | Understand system behavior | Which parameters matter most |
| **Optimization** | Find best management strategy | Optimal pumping rates |
| **Risk assessment** | Evaluate probabilities | Chance of contamination reaching well |

### Before You Apply the Model

Before running prediction scenarios, ask yourself:

1. **What question am I trying to answer?** (Be specific)
2. **Is my model capable of answering this question?** (Appropriate complexity, validated domain)
3. **What are the key uncertainties?** (From Notebook 7)
4. **How will the results be used?** (Screening vs. detailed design)

### The Danger of Misapplication

A common mistake is using models beyond their capabilities:

- Using a **steady-state model** to predict transient responses
- Making predictions **outside the calibrated domain**
- Ignoring **structural uncertainty** in model design
- Treating model outputs as **exact values** rather than estimates

> **Key Principle**: A model should only be used for purposes consistent with its design, calibration, and validation status.

## 2 - Defining Prediction Scenarios

### Elements of a Well-Defined Scenario

Every prediction scenario should clearly specify:

| Element | Question to Answer | Example |
|---------|-------------------|----------|
| **Objective** | What are we trying to predict? | Maximum drawdown at supply well |
| **Spatial scope** | Where are predictions needed? | Within 500 m of proposed well |
| **Temporal scope** | What time period? | First 5 years of operation |
| **Boundary conditions** | What external conditions apply? | Average hydrological year |
| **Stresses** | What changes are we simulating? | New well pumping at 1000 m³/day |
| **Outputs** | What results do we need? | Head contours, drawdown cone |

### Types of Scenarios

#### Baseline Scenario
Current or historical conditions—used as reference for comparison.

**Purpose**: Establish "what is" before evaluating "what if"

#### Stress Testing Scenarios
Push the system to understand limits and vulnerabilities.

**Examples**:
- Drought conditions (reduced recharge)
- Maximum permitted pumping
- Worst-case contamination release

#### Future Condition Scenarios
Project how conditions may change over time.

**Examples**:
- Climate change impacts on recharge
- Population growth affecting demand
- Land use change affecting infiltration

#### Management Scenarios
Evaluate different intervention strategies.

**Examples**:
- Alternative well configurations
- Remediation designs
- Managed aquifer recharge options

### Temporal Considerations

| Model Type | Appropriate Predictions | Inappropriate Predictions |
|------------|------------------------|---------------------------|
| **Steady-state** | Long-term average conditions | Response times, recovery periods |
| **Transient** | Time-varying responses | Conditions beyond simulation period |
  

> **Our Model**: The Limmat Valley model is steady-state, so it can predict equilibrium conditions but not how quickly changes occur.

## 3 - Model Limitations for Specific Applications

### Matching Model Capabilities to Needs

Before applying a model, honestly assess whether it's fit for purpose:

```
┌─────────────────────────────────────────────────────────────┐
│  APPLICATION REQUIREMENTS   →   MODEL CAPABILITIES         │
├─────────────────────────────────────────────────────────────┤
│  What predictions are needed?                               │
│           ↓                                                 │
│  Can the model physics represent the relevant processes?    │
│           ↓                                                 │
│  Has the model been validated for this type of prediction?  │
│           ↓                                                 │
│  Are uncertainties acceptable for the decision at hand?     │
└─────────────────────────────────────────────────────────────┘
```

### Our Limmat Valley Model: Honest Assessment

From Notebooks 5-7, we know our model has specific limitations:

| Characteristic | Implication for Application |
|----------------|----------------------------|
| **Steady-state** | Cannot predict timing of changes |
| **3 observation wells** | Limited validation, unverified in most areas |
| **Manual calibration** | Parameter uncertainty not formally quantified |
| **Single layer** | Cannot resolve vertical flow components |
| **Uniform K zones** | Does not capture local heterogeneity |

### What Our Model CAN and CANNOT Do

| ✅ Appropriate Uses | ❌ Inappropriate Uses |
|---------------------|----------------------|
| Screening-level assessment | Regulatory permit applications |
| Conceptual understanding of flow patterns | Detailed well placement optimization |
| Identifying data gaps | Precise drawdown predictions |
| Teaching groundwater concepts | Contaminant arrival time predictions |
| Preliminary scenario comparison | Legal or financial decisions |

### When to Refine vs. Accept Limitations

| Decision Context | Recommended Action |
|------------------|-------------------|
| Low-stakes screening | Accept limitations, document caveats |
| Moderate consequences | Add safety factors, consider additional data |
| High-stakes decisions | Refine model, collect more data, quantify uncertainty |
| Regulatory submission | Full validation, uncertainty analysis required |

### Documenting Assumptions and Caveats

Every model application should document:

1. **Model version and calibration status**
2. **Key assumptions** (boundary conditions, parameter values)
3. **Known limitations** relevant to the application
4. **Uncertainty bounds** on predictions (if available)
5. **Recommended uses** of the results

## 4 - Scenario Design Principles

### Principle 1: Physically Plausible Parameter Ranges

Scenarios should test conditions that could actually occur:

| Parameter | Reasonable Range | Unreasonable |
|-----------|-----------------|---------------|
| Recharge | ±50% of average | 10× increase |
| Pumping | Up to permitted rate | Unlimited extraction |
| River stage | Historical range | Arbitrary values |

**Why it matters**: Predictions from unrealistic scenarios have no practical value.

### Principle 2: Consistent Boundary Conditions

When modifying stresses, consider effects on boundaries:

- If you increase pumping, do river levels stay the same?
- If climate changes, do constant head boundaries still apply?
- Are boundary conditions consistent with the scenario being tested?

**Common mistake**: Changing internal stresses while ignoring boundary effects.

### Principle 3: Fair Comparison Between Scenarios

To compare scenarios meaningfully:

1. **Same baseline**: All scenarios start from the same calibrated state
2. **Controlled changes**: Only change variables relevant to the question
3. **Same metrics**: Compare scenarios using consistent output measures
4. **Same uncertainty treatment**: Apply same caveats to all scenarios

### Principle 4: Bracket Uncertainty

Rather than running a single "best estimate" scenario:

- Run **low**, **medium**, and **high** cases
- Vary the most sensitive and uncertain parameters (from Notebook 7)
- Report range of outcomes, not single values

### Principle 5: Consider Extreme Conditions

Models are often most useful for stress testing:

| Stress Type | Scenario Example |
|-------------|------------------|
| Drought | 50% reduction in recharge for 3 years |
| Flood | River stage at 100-year flood level |
| Contamination | Spill at worst-case location |
| Over-pumping | All wells at maximum capacity |

**Purpose**: Understand system vulnerability, not predict exact outcomes.

## 5 - Interpreting and Using Results

### Predictions Are Not Facts

Model outputs should never be presented as certainties:

| ❌ Poor Communication | ✅ Better Communication |
|-----------------------|------------------------|
| "The drawdown will be 2.5 m" | "The model predicts drawdown of 2-3 m" |
| "The plume will arrive in 2027" | "Travel time is estimated at 5-10 years" |
| "The design is safe" | "Under tested conditions, heads stay above minimum" |

### Linking to Uncertainty Analysis

From Notebook 7, recall that predictions carry uncertainty from:

- Parameter uncertainty (K, recharge, etc.)
- Model structural uncertainty (simplifications)
- Future scenario uncertainty (boundary conditions)

**Best practice**: Report predictions with confidence intervals when possible.

### Decision-Making Under Uncertainty

Different decisions require different confidence levels:

| Decision Type | Uncertainty Tolerance | Example |
|--------------|----------------------|----------|
| Screening | High | "Worth investigating further?" |
| Planning | Moderate | "What capacity should we plan for?" |
| Design | Low | "Will this well meet water quality standards?" |
| Regulatory | Very low | "Will this remediation achieve cleanup goals?" |

### When Results Are "Good Enough"

A prediction is sufficient when:

1. The **uncertainty range** doesn't change the decision
2. **Conservative assumptions** provide adequate safety margin
3. The **consequences of being wrong** are acceptable
4. **Additional refinement** wouldn't change the outcome

**Example**: If all scenarios predict the contaminant stays within property boundaries, the exact plume extent may not matter for the decision.

### Documenting Results for Future Use

Model applications should be documented so results can be:

- **Reproduced** by others
- **Updated** when new data becomes available
- **Compared** with actual outcomes (model validation)
- **Cited** in reports and decisions

## 6 - Example Application Considerations

Let's consider how we would approach two hypothetical applications using the Limmat Valley model.

### Example 1: Impact of a New Pumping Well

**Question**: What would be the impact of installing a new pumping well for industrial cooling?

#### Considerations Before Running the Model

| Consideration | Our Assessment |
|---------------|----------------|
| **Can our model answer this?** | Partially—we can estimate equilibrium drawdown, but not response time |
| **Is the location validated?** | Only if near our 3 observation wells; otherwise, predictions are extrapolations |
| **What parameters matter?** | K (controls drawdown), river leakance (controls boundary response) |
| **What outputs are needed?** | Drawdown at well, impact on nearby wells, effect on river |

#### Scenario Design

1. **Baseline**: Current conditions (calibrated model)
2. **Proposed scenario**: Add well at specified location and rate
3. **Sensitivity scenarios**: Vary K (±20%) to bracket uncertainty
4. **Stress test**: Maximum expected pumping rate

#### Caveats to Report

- Steady-state model—cannot predict how quickly drawdown develops
- Model not validated near proposed location—predictions are indicative only
- Recommend monitoring program if well is installed

---

### Example 2: Contaminant Transport Prediction

**Question**: If a spill occurred at an industrial site, how long would it take to reach the Limmat River?

#### Considerations Before Running the Model

| Consideration | Our Assessment |
|---------------|----------------|
| **Can our model answer this?** | Limited—flow model provides advective travel time, but no dispersion or retardation |
| **What additional model would be needed?** | MT3DMS or similar transport model |
| **What parameters are most uncertain?** | Porosity, dispersivity, sorption (not calibrated) |
| **Is the flow path validated?** | Partially—general direction is consistent with expectations |

#### What We Could Do

1. Calculate **advective travel time** using Darcy velocity and porosity
2. Apply **range of porosities** (0.1-0.3) to bracket travel time
3. Acknowledge this is a **screening estimate** only

#### Caveats to Report

- Advective-only estimate ignores dispersion and retardation
- Actual arrival could be faster (preferential pathways) or slower (sorption)
- Recommend site-specific investigation if this is a real concern

## 7 - Advanced Topics

For professional practice and advanced research, more sophisticated application methods are available.

### Optimization and Management Models

Coupling groundwater models with optimization algorithms to find best management strategies.

**Applications**:
- Optimal well placement for maximum yield with minimum interference
- Pumping schedules that minimize energy cost while meeting demand
- Remediation system design (e.g., optimal pump-and-treat configuration)

**Methods**:
- Linear programming for simple constraints
- Genetic algorithms for complex, nonlinear problems
- Response surface methods for computationally expensive models

**Tools**: MODFLOW with PEST++ optimization, GWM (Groundwater Management Process)

### Stochastic Modeling

Using multiple equally-likely parameter realizations to represent heterogeneity.

**Concept**: Instead of a single "best estimate" K field, generate 100+ realizations that all honor available data. Run model for each realization.

**Output**: Probability distributions of predictions (e.g., "80% of realizations show plume staying within 500 m")

**When to use**: When heterogeneity significantly affects predictions, especially for contaminant transport.

**Tools**: TPROGS (transition probability geostatistics), PEST++ IES for ensemble-based analysis

### Model Coupling

Linking groundwater models with other environmental models.

| Coupling Type | Example Application |
|---------------|--------------------|
| **Surface water** | Stream-aquifer interaction, conjunctive use |
| **Land surface** | Evapotranspiration, soil moisture |
| **Climate models** | Long-term recharge projections |
| **Land use models** | Urbanization impacts on infiltration |
| **Economic models** | Cost-benefit analysis of water management |

**Tools**: GSFLOW (MODFLOW + PRMS), ParFlow, integrated modeling frameworks

### Transient Predictions and Climate Change

Moving beyond steady-state to time-varying scenarios.

**Considerations**:
- How will seasonal patterns change?
- What are the impacts of more frequent droughts?
- How do storage changes affect long-term sustainability?

**Challenge**: Uncertainty compounds over longer prediction horizons.

### Decision Support Under Deep Uncertainty

When probabilities cannot be reliably estimated (e.g., future climate, policy changes).

**Approaches**:
- **Robust Decision Making (RDM)**: Find strategies that perform well across many futures
- **Info-Gap Theory**: Maximize robustness to uncertainty
- **Scenario Planning**: Test performance under distinct plausible futures

**Key insight**: Instead of optimizing for one predicted future, find strategies that are acceptable across many possible futures.

### Real-Time Modeling and Data Assimilation

Continuously updating model predictions as new observations become available.

**Applications**:
- Flood forecasting with real-time river stage data
- Contaminant tracking with ongoing monitoring
- Adaptive management of pumping systems

**Methods**:
- Kalman filtering for sequential updates
- Ensemble methods (PEST++ IES) for batch updates
- Machine learning for rapid response

## 8 - Summary and Checklist

### Key Messages

1. **Define questions clearly**: Know exactly what you're trying to predict before running scenarios

2. **Match capabilities to needs**: Only use the model for purposes consistent with its design and validation

3. **Design scenarios thoughtfully**: Use physically plausible parameters and consistent boundary conditions

4. **Report with uncertainty**: Predictions are estimates, not facts—communicate ranges and caveats

5. **Document everything**: Enable reproducibility and future updates

### Pre-Application Checklist

Before running prediction scenarios, verify:

| Item | Status |
|------|--------|
| □ **Objective clearly defined** | What question are we answering? |
| □ **Model capable of prediction** | Does model physics support this application? |
| □ **Prediction within validated domain** | Has model been tested for similar predictions? |
| □ **Scenarios physically plausible** | Are parameter changes realistic? |
| □ **Boundary conditions consistent** | Do boundaries reflect scenario conditions? |
| □ **Uncertainty considered** | Have we bracketed key uncertain parameters? |
| □ **Caveats documented** | Are limitations clearly stated? |
| □ **Results reproducible** | Can someone else recreate this analysis? |

### Model Use Categories (from Notebook 6)

| Validation Status | Appropriate Uses | Inappropriate Uses |
|-------------------|------------------|--------------------|
| **Unvalidated** | Screening, conceptual understanding | Regulatory decisions, design |
| **Partially validated** | Preliminary design | Final design, permits |
| **Fully validated** | Regulatory submissions, detailed predictions | None (but still report uncertainty) |
