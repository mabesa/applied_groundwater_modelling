# Theory Reminder Page

## Flow

### Main variables and soil properties
- Discharge, $Q$, in L/s
- Porosity, $\phi$, no unit.
- Water content, $\Theta_w$, not unit.
- Saturation water content, $S_w$, no unit.


### Bernoulli law
- Hydraulic head, $h$, in m

$h = z + \frac{p}{\rho g}$



### Darcy's law
- Specific discharge, $\vec{q}$, in m $\text{s}^{-1}$
- Discharge, $\vec{Q}$, i $\text{m}^{3}\text{s}^{-1}$
- Aquifer cross section $A$ in $\text{m}^{2}$
- Hydraulic conductivity tensor $-\mathbf{K}$ in m $\text{s}^{-1}$
- Hydraulic gradient $I$

$ \vec{q} = \frac{\vec{Q}}{A} 
            = - \mathbf{K} \cdot \frac{\Delta h}{L} 
            = - \mathbf{K} \vec{I} 
            = - \mathbf{K} \cdot \vec{\nabla h}$
            

### Equivalent hydraulic conductivity
Horizontally superimposed layers :

$K_{eq} =  \frac{1}{d} \Sigma d_i K_i$

Vertically superimposed layers : 

$\frac{1}{K_{eq}} =  \frac{1}{d} \Sigma d_i \frac{1}{K_i}$

### Transmissivity

Confined layers : \mathbf{T} = \mathbf{K} \Delta m

Unconfined layer : \mathbf{T} = \mathbf{K} h(x,y) - z_{bottom}


### Mass balance equation
- Specific discharge, $\vec{q}$, in m $\text{s}^{-1}$
- Specific storativity $S_0$, 
- Storativity $S$, in $\text{m}^{-1}$
- Aquifer compressibility $\alpha_{aq}$
- Water compressibility $\beta_{w}$
- Aquifer drainable porosity $\phi_d$

$\partial_t(\rho \phi) = - \nabla \cdot \rho \vec{q} + \rho w = \rho S_0 \partial_t h + \rho w $

$S_0 = \rho g (\partial_p \phi + \phi \frac{\partial_p \rho}{\rho})
            = \rho g (\alpha_{aq} + \phi \beta_{w})$

$S = S_0 m$ in a confined aquifer

$S = \phi_d $ in an unconfined aquifer


### Formulation of flow problem 

$S_0 \partial_t h = \partial_x K \partial_x h + \partial_y K \partial_y h + \partial_z K \partial_z h + W $

- Source or sinks, W, possibly dicharge from :

recharge rate $N$

lateral inflow $q_l$

well input $-\frac{Q_{well}}{A_{well}}$

river leakage 

### Boundary conditions for flow problem

Specified constant head, (eg. a river) : $h(x,y,t) = h_{riv}(x,y,t)$

Specified recharge flow (in or out) :  $Q_z = N_z \Delta x \Delta y$

Impermeable boundary : $\frac{\partial h}{\partial n}=0$


### Simple analytic solutions to flow problem

Steady state : \partial_t h
Well radius of influence : $R(t) = 1.5 \sqrt{\frac{T}{S}t}$

#### Confined 1D flow : 
Flow problem : $S_0 \cdot 0 = \partial_x K \partial_x h + 0 + 0 + 0 $

$h(x) = h(x=0) + \frac{h(x=L)-h(x=0)}{L} \cdot x$

$Q = B_y T \frac{h(x=L)-h(x=0)}{L}$

#### Unconfined 1D flow :
Flow problem : $S_0 \cdot 0 = \partial_x K \partial_x h + 0 + 0 + N $

$h(x)^2 = h(x=0)^2 - (h(x=0)^2 - h(x=L)^2)\frac{x}{L} + \frac{N}{K} \cdot x (x-L)$ (Dupuit's solution)

$Q(x) = \frac{B_y K (h(x=0)^2 - h(x=L)^2)}{2L} + NB_y (x-\frac{L}{2})$

#### Confined 2D radial flow :
Flow problem : $r\partial_rh = -\frac{Q_w}{2 \pi T}$

$h(r)= h(r=0) - \frac{Q_w}{2 \pi T} \ln (\frac{r}{R(t)})$



#### Unconfined 2D radial flow :
$h(r)^2= h(r=0)^2 - \frac{Q_w}{K \pi} \ln (\frac{r}{R(t)})$





## Transport

### Transport equation

$\partial_t c_{tot} (x,y,z,t) = - \Delta \cdot \vec{j}(x,y,z,t)+P(x,y,z,t)$

$\partial_t(\phi_e R c) = - \vec{\nabla} \cdot (\phi_e \vec{u} c)
                            + \vec{\nabla} \phi_e (D_{me} + \bold{D}) \vec{\nabla c}
                            - \lambda \phi_e R c$

For constant specific velocity and porosity : $\partial_t( R c) = -  \vec{u} \vec{\nabla} c
                            + (D_{me} + \bold{D})
                            + \vec{\nabla}^2 c
                            - \lambda R c$

### Advection mass flux

- Seepage velocity, $\vec{u}$, in  m $\text{s}^{-1}$

$\vec{j_a} = \phi_e \vec{u} c = \vec{q} c$



### Molecular diffusion mass flux

- Dispersivity $\bold{D_{me}}$ is isotropic and equals $10^{-9}$ in all directions, in  $\text{m}^{2}\text{s}^{-1}$

$\vec{j_m} = - \phi_e \bold{D_{me}} \vec{\nabla}c$

### Dispersion mass flux

- Dispersivity $\bold{D}$ is such that: 
    - $D_{ij} = \frac{1}{|u|}(u_i u_j(\alpha_L-\alpha_T))$
    - $D_{ii} = \frac{1}{|u|}(\alpha_T(u_j^2+u_k^2)+ \alpha_L u_i^2)$

$\vec{j_d} = - \phi_e \bold{D} \vec{\nabla}c$
