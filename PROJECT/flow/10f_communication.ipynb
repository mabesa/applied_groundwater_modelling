## 1 — Know Your Audience

Different audiences need different levels of detail. The same model result must be translated for each:

| Audience | Needs | Avoid |
|----------|-------|-------|
| **Policymakers** | Key findings, implications, recommendations | Jargon, equations |
| **Regulators** | Compliance, uncertainty bounds, data quality | Oversimplification |
| **Engineers** | Design parameters, safety factors | Excessive theory |
| **Scientists** | Methods, assumptions, limitations | Missing details |
| **Public** | Plain language, relevance to daily life | Any jargon |

### 1.1 - The "So What?" Test

For every result, ask: *"So what does this mean for this audience?"*

| Our model result | For a policymaker | For an engineer |
|-----------------|-------------------|-----------------|
| K = 5–50 m/d (spatially variable) | "Groundwater flows fast — contaminants can spread quickly" | "Use the calibrated K field; expect 5–50 m/d depending on location" |
| RMSE ~1 m (NB5/NB6) | "Model predictions match observations within about 1 metre" | "Design tolerance ±1 m on predicted heads" |
| S2 zone extends ~100–300 m (NB8) | "A protection zone this size restricts land use around the well" | "Use 300 m as conservative S2 boundary for design" |
| High K uncertainty in western domain (NB7) | "Results in the western area are less reliable — we need more data" | "Apply larger safety factors west of the observation cluster" |

### 1.2 - Layered Communication

Structure reports so each reader finds their level:

1. **Executive summary** (1 page) — key findings and recommendations
2. **Main report** (10–20 pages) — methods, results, discussion
3. **Technical appendices** — full details for specialists

## 2 — Effective Visualizations

### 2.1 - Five Principles

1. **One message per figure** — don't overload
2. **Clear labels** — axes, units, legends, colorbars
3. **Appropriate scale** — show the relevant variation, not the full data range
4. **Accessible colours** — use colourblind-safe palettes (viridis, RdBu); avoid rainbow
5. **Stand-alone captions** — a reader should understand the figure without reading the full text

### 2.2 - Which Plot for Which Purpose?

| Purpose | Visualization | Our example |
|---------|--------------|-------------|
| Spatial distribution | Head / drawdown map | NB4 head map, NB8 drawdown cone |
| Model fit | Observed vs. simulated scatter | NB5 calibration scatter |
| Uncertainty | Shaded ranges, box plots | NB7 FOSM confidence intervals |
| Water balance | Bar chart or table | NB8 baseline vs. pumping budget |
| Flow paths | Pathline map | NB8 PRT capture zone |
| Sensitivity | Tornado or bar chart | NB7 parameter sensitivity |
| Conceptual overview | Cross-section diagram | NB2 perceptual model |

### 2.3 - Audience-Specific Choices

**Technical audiences** want detail: coordinate grids, statistics (RMSE, bias), multi-panel comparisons, log-scale axes where appropriate.

**Non-technical audiences** want context: annotated maps with landmarks and street names, simplified conceptual diagrams, infographics with key numbers highlighted. Replace "Head = 402 m a.s.l." with "Groundwater sits about 8 m below the surface here."

### 2.4 - Colour Guidance

| Data type | Use | Avoid |
|-----------|-----|-------|
| Sequential (heads, K) | `viridis`, `plasma` | Rainbow (perceptually uneven) |
| Diverging (drawdown, residuals) | `RdBu`, `coolwarm` | Green–red (colourblind-unsafe) |
| Categorical (zones, wells) | `Set2`, `tab10` | Too many colours (>7) |

## 3 — Communicating Uncertainty

All models are uncertain. Hiding uncertainty erodes trust, leads to poor decisions, and is professionally unethical. The goal is to **communicate uncertainty honestly without paralysing the audience**.

### 3.1 - Ways to Express Uncertainty

| Method | Example from our model | Best for |
|--------|----------------------|----------|
| **Ranges** | "The S2 zone extends 100–300 m from the well" | Simple communication |
| **Confidence intervals** | "Head at well W3: 402 ± 2 m (95% CI from FOSM)" | Technical reports |
| **Scenarios** | "Under drought, the capture zone expands by ~30%" | Decision support |
| **Qualitative** | "High confidence near observation wells, low confidence in the west" | Non-technical audiences |

### 3.2 - Sources of Uncertainty to Report

From our Limmat Valley modelling experience:

| Source | Where we encountered it | How to communicate |
|--------|------------------------|-------------------|
| **Parameter uncertainty** | K varies 5–50 m/d; FOSM shows wide bounds in western domain (NB7) | Maps with confidence shading |
| **Structural uncertainty** | Single layer vs. multi-layer; steady-state vs. transient | State assumptions and their implications |
| **Data gaps** | Only 4 real observation wells; none in the western third | "Model is unconstrained west of [location]" |
| **Scenario uncertainty** | Drought recharge reduction of 30% — is that realistic? | Present as "what-if", not prediction |

### 3.3 - Honest Reporting Principles

**Do:** State what is known vs. assumed. Provide ranges, not point estimates. Put limitations in the executive summary.

**Don't:** Use false precision ("arrives in 7.34 years"). Hide uncertainty in appendices. Claim the model "proves" anything — models provide evidence, not proof.

> **📚 Classifying Model Confidence:**
> When communicating model results to stakeholders, it helps to classify the model's confidence level using a recognised framework. The **DVGW W 107** standard (used in German-speaking countries) and the **Australian Table 2-1** (Barnett et al., 2012) both provide tiered classifications linking data quality, calibration status, and validation to appropriate model uses. A Class 1 model (limited data, little calibration) should only support conceptual understanding, while a Class 3 model (extensive data, comprehensive validation) can inform regulatory decisions. Presenting your model's classification explicitly helps stakeholders understand which decisions the model can and cannot support. See also the classification exercise in [Notebook 6](06f_validation.ipynb).

## 4 — Common Pitfalls

Five mistakes that undermine even good models:

### Pitfall 1: False Precision

| Wrong | Right |
|-------|-------|
| "Drawdown at the well is 1.847 m" | "Drawdown is approximately 1–3 m, depending on local K" |

*Why:* Our K field has high uncertainty in the western domain (NB7). Three decimal places imply a confidence the model doesn't have.

### Pitfall 2: Buried Limitations

| Wrong | Right |
|-------|-------|
| Limitations in Appendix C, footnote 12 | Limitations in the executive summary, paragraph 2 |

*Why:* Decision-makers read the summary. If they don't see "model is unvalidated in the western domain," they'll treat all predictions equally.

### Pitfall 3: Jargon Overload

| Wrong | Right |
|-------|-------|
| "RMSE of 0.8 m indicates acceptable calibration per ASTM guidelines" | "Model predictions match observed water levels to within about 1 metre" |

*Why:* Most readers aren't groundwater specialists. Translate metrics into plain language.

### Pitfall 4: Numbers Without Context

| Wrong | Right |
|-------|-------|
| "Heads range from 392 to 410 m" | "Groundwater levels drop 18 m along the valley, from 410 m near Höngg to 392 m at the Limmat outlet — confirming flow follows the river" |

*Why:* Numbers need interpretation. Link them to geography and physical meaning.

### Pitfall 5: Overconfidence

| Wrong | Right |
|-------|-------|
| "The model proves contamination will not reach the well" | "Under the tested scenarios, contamination is unlikely to reach the well within 10 days. This depends on K values that are uncertain in this area (NB7)" |

*Why:* Models cannot "prove" anything. They provide evidence under stated assumptions.

## 5 — Report Structure (Standard Outline)


```
1. Executive Summary (1–2 pages)
   Key findings, recommendations, major limitations
```

The executive summary is often the only part read. Make it count!
1. **Lead with the answer** — "The proposed well causes ~1–3 m drawdown; the river supplies the pumped water"
2. **Be specific** — numbers, not vague statements
3. **Include limitations** — "Model is unvalidated in the western domain"
4. **End with recommendations** — "Install two additional monitoring wells before finalising the protection zone"

```
2. Introduction
   Objectives, site description, scope

3. Data and Methods
   Data sources and quality, model setup, calibration approach

4. Results
   Calibration performance, scenario results, sensitivity

5. Discussion
   Interpretation, uncertainty, comparison with expectations

6. Conclusions and Recommendations
   Key findings (bulleted), actionable next steps, data gaps

7. References

Appendices
   A. Detailed parameter tables
   B. Additional figures
   C. Model archive location
```
