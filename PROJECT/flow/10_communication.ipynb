## 1 - Know Your Audience

### Different Audiences, Different Needs

| Audience | Needs | Avoid |
|----------|-------|-------|
| **Policymakers** | Key findings, implications, recommendations | Technical jargon, equations |
| **Regulators** | Compliance, uncertainty bounds, data quality | Oversimplification |
| **Engineers** | Design parameters, safety factors | Excessive theory |
| **Scientists** | Methods, assumptions, limitations | Missing details |
| **Public** | Plain language, relevance to them | Any jargon |

### The "So What?" Test

For every result, ask: **"So what does this mean for the audience?"**

| Technical Finding | For Policymaker | For Engineer |
|-------------------|-----------------|---------------|
| "K = 17 m/day" | "Groundwater flows relatively fast" | "Use K = 17 m/day in design" |
| "RMSE = 0.8 m" | "Model matches observations well" | "Expect ±1m accuracy" |
| "Model unvalidated" | "Results are screening-level only" | "Add safety factors" |

### Layered Communication

Structure reports with increasing detail:

1. **Executive Summary** (1 page) - Key findings and recommendations
2. **Main Report** (10-20 pages) - Methods, results, discussion
3. **Technical Appendices** - Full details for specialists

## 2 - Effective Visualizations

### Principles of Good Visualization

1. **One message per figure** - Don't overload
2. **Clear labels** - Axes, units, legends
3. **Appropriate scale** - Show relevant variation
4. **Accessible colors** - Consider colorblindness
5. **Caption explains** - Figure should stand alone

### Visualization Types for Groundwater

| Purpose | Visualization | When to Use |
|---------|--------------|-------------|
| Spatial distribution | Head contour map | Show flow patterns |
| Model fit | Scatter plot (obs vs sim) | Calibration results |
| Time series | Hydrograph | Transient results |
| Uncertainty | Box plots, error bars | Parameter ranges |
| Water balance | Bar chart, Sankey diagram | Budget components |
| Sensitivity | Tornado diagram | Parameter importance |
| Conceptual | Cross-section diagram | Explain system |

### Audience-Specific Visualizations

**For Technical Audiences:**
- Detailed contour maps with coordinates
- Scatter plots with statistics
- Multiple panels showing sensitivity

**For Non-Technical Audiences:**
- Simplified conceptual diagrams
- 3D visualizations for spatial understanding
- Annotated maps with landmarks
- Infographics summarizing key points

### Color Considerations

| Use | Recommended | Avoid |
|-----|-------------|-------|
| Sequential data | viridis, plasma | Rainbow |
| Diverging data | RdBu, coolwarm | Green-red |
| Categorical | Set2, tab10 | Too many colors |

## 3 - Communicating Uncertainty

### Why Communicate Uncertainty?

All models are uncertain. Hiding uncertainty:
- Erodes trust when predictions fail
- Leads to poor decisions
- Is professionally unethical

### Ways to Express Uncertainty

| Method | Example | Best For |
|--------|---------|----------|
| **Ranges** | "5-10 years" | Simple communication |
| **Confidence intervals** | "402 ± 2 m (95% CI)" | Technical reports |
| **Probability statements** | "80% chance of arrival by 2030" | Risk communication |
| **Scenarios** | "Best/worst/likely case" | Decision support |
| **Qualitative** | "High/medium/low confidence" | Non-technical audiences |

### Uncertainty Visualization

| Type | Visualization |
|------|---------------|
| Parameter ranges | Box plots, violin plots |
| Spatial uncertainty | Maps with confidence bounds |
| Temporal uncertainty | Shaded envelopes on hydrographs |
| Scenario comparison | Side-by-side maps, bar charts |

### Honest Uncertainty Communication

**Do:**
- State what is known vs. assumed
- Explain sources of uncertainty
- Provide ranges, not just point estimates
- Note model limitations prominently

**Don't:**
- Hide uncertainty in appendices
- Use false precision ("arrives in 7.3 years")
- Overstate confidence
- Ignore structural uncertainty

## 4 - Report Structure

### Standard Technical Report Outline

```
1. Executive Summary (1-2 pages)
   - Key findings
   - Recommendations
   - Major limitations

2. Introduction
   - Project objectives
   - Site description
   - Report scope

3. Data and Methods
   - Data sources and quality
   - Model setup (conceptual → numerical)
   - Calibration approach

4. Results
   - Calibration performance
   - Scenario results
   - Sensitivity analysis

5. Discussion
   - Interpretation
   - Uncertainty and limitations
   - Comparison with expectations

6. Conclusions and Recommendations
   - Key findings (bulleted)
   - Actionable recommendations
   - Data gaps and future work

7. References

Appendices
   A. Detailed methods
   B. Model parameters
   C. Additional figures
```

### Executive Summary Best Practices

The executive summary is often the only part read. Make it count:

1. **Lead with the answer** - What did you find?
2. **Be specific** - Numbers, not vague statements
3. **Include limitations** - Don't oversell
4. **End with recommendations** - What should be done?

### Tables for Key Results

Use tables to summarize clearly:

| Scenario | Travel Time | Confidence | Notes |
|----------|-------------|------------|-------|
| Base case | 8 years | Medium | Current conditions |
| Drought | 12 years | Low | 50% recharge reduction |
| High pumping | 5 years | Medium | Maximum permitted rate |

## 5 - Common Pitfalls

### Pitfall 1: False Precision

**Wrong:** "The contaminant will arrive in 7.34 years"

**Right:** "The contaminant is expected to arrive in 5-10 years"

*Why:* Model uncertainty doesn't support decimal precision.

### Pitfall 2: Buried Limitations

**Wrong:** Limitations mentioned only in appendix footnotes

**Right:** Limitations stated in executive summary and discussion

*Why:* Decision-makers need to know what the model cannot do.

### Pitfall 3: Jargon Overload

**Wrong:** "The RMSE of 0.8m indicates acceptable calibration per ASTM guidelines"

**Right:** "The model predictions match observed water levels to within about 1 meter, which is considered acceptable for this type of analysis"

*Why:* Most readers aren't groundwater specialists.

### Pitfall 4: Missing Context

**Wrong:** "Heads range from 395 to 410 m"

**Right:** "Groundwater levels range from 395m near the river to 410m at the valley edge, indicating flow toward the river"

*Why:* Numbers need interpretation.

### Pitfall 5: Overconfidence

**Wrong:** "The model proves that contamination will not reach the well"

**Right:** "Under the scenarios tested, contamination is unlikely to reach the well within 20 years. However, this conclusion depends on assumptions about [X, Y, Z] that carry uncertainty"

*Why:* Models cannot "prove" anything - they provide evidence.

## 6 - Advanced Topics

### Interactive Visualizations

Web-based tools for stakeholder engagement:

- **Panel/Holoviews**: Interactive Python dashboards
- **Plotly**: Interactive plots embeddable in reports
- **Model Viewer**: 3D model visualization (USGS)
- **Web mapping**: Leaflet, Folium for spatial results

### Structured Decision Making

Framework for presenting results to support decisions:

1. Define objectives and constraints
2. Identify alternatives
3. Evaluate consequences (model results)
4. Assess trade-offs
5. Make recommendation

### Risk Communication

Principles from risk communication research:

- **Comparisons help** - "Similar to..." makes abstract risks concrete
- **Frequency better than probability** - "1 in 100 years" vs. "1% chance"
- **Acknowledge uncertainty** - Increases, not decreases, trust
- **Two-way communication** - Allow questions and feedback

### Reproducible Reports

Generate reports directly from code:

- **Jupyter notebooks** → PDF/HTML reports
- **Quarto/R Markdown** → Dynamic documents
- **Automated figure generation** → Consistent updates

### Model Summary Sheets

One-page summaries for quick reference:

```
┌────────────────────────────────────────┐
│ MODEL SUMMARY: Limmat Valley v1.2     │
├────────────────────────────────────────┤
│ Purpose: Screening-level flow analysis │
│ Status: Calibrated, not validated      │
│ Key result: GW flows toward Limmat     │
│ Limitation: 3 obs wells only           │
│ Contact: [name], [date]                │
└────────────────────────────────────────┘
```

## 7 - Summary

### Communication Checklist

Before sharing results:

| Item | Check |
|------|-------|
| ☐ Audience identified | Who will read this? |
| ☐ Key message clear | What's the "so what"? |
| ☐ Uncertainty stated | Ranges, not false precision |
| ☐ Limitations prominent | In summary, not just appendix |
| ☐ Visualizations clear | One message per figure |
| ☐ Jargon minimized | Would a non-specialist understand? |
| ☐ Recommendations actionable | What should be done? |

### Key Messages

1. **Know your audience** - Tailor detail and language
2. **Lead with findings** - Don't bury the answer
3. **Visualize effectively** - One clear message per figure
4. **Communicate uncertainty** - Honestly and prominently
5. **State limitations** - Builds trust, not weakness
6. **Make it actionable** - End with recommendations

### The Bottom Line

> A model that isn't communicated effectively might as well not exist.
> Your analysis is only as valuable as your ability to convey its meaning.
