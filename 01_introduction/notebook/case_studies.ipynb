# Introducing : The Case Studies

## Case Study 1 : Infrastructure Projects in Zurich, Switzerland

In this case study, you step into the role of an Engineering geologist at a major Swiss consulting firm, specializing in complex urban construction projects in Zurich. You are responsible for ensuring safe construction while protecting groundwater resources and managing various geological challenges, which are: 

### Limmat Valley High-Rise
A 120m high-rise planned in the Limmat Valley with 4 underground parking levels. The excavation will reach 15m below groundwater level in the highly permeable Limmat Valley aquifer.

Technical Challenges:

- Estimate of groundwater inflow rates  
- Evaluation of groundwater drawdown impacts   

Learning Objectives: Darcy equation, aquifer properties, numerical modeling

### Uetliberg Slope Engineering
Design of spring catchment and water reservoir on the unstable, mergeliger (marly) slope of Uetliberg.

Technical Challenges:

- Spring protection zones design  
- Drainage system planning  

Learning Objectives: Groundwater flow in fractured media, protection zones

### Thermal Use Conflict
Multiple groundwater heat pump systems planned in a densely built area.

Technical Challenges:

- Thermal plume modeling  
- System interference assessment  
- Sustainable extraction rates  
- Temperature monitoring design  

Learning Objectives: Heat transport, thermal plume modeling

### Historic Industrial Contamination
Former industrial site shows chlorinated hydrocarbon contamination threatening downstream drinking water wells.

Technical Challenges:

- Contaminant transport modeling  
- Risk assessment  
- Remediation strategy  
- Construction adaptation  

Learning Objectives: Contaminant transport, remediation concepts

- Well location optimization
